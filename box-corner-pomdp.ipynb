```python
class PlanarPolygonObjectInCorner():
    def __init__(self,vertex_list=None):
        
        if vertex_list is None:
            vertex_list = [ [1,1], [1,-1], [-1,-1], [-1,1] ]
      
        self.num_vertices = len(vertex_list)
        #self.edge_list = [(i,i+i) for i in range(self.num_vertices-1)] + [(self.num_vertices-1,0)]
    
        self.vertex_list_original = np.copy(vertex_list)
        
        self.wall_x = 0
        self.wall_y = 0
        
        self.vertex_list = None
        self.pose = None
        
        self.restore()


    def set_pose(self,pose):
        self.restore()
        self.move(*pose)
        
    def get_pose(self):
        return self.pose  
        
    def move(self,x,y,theta):
        rot = [[np.cos(theta), np.sin(theta)],
               [-np.sin(theta), np.cos(theta)]]
        
        xy_c = self.get_pose()[0:2]
        self.vertex_list = np.dot(self.vertex_list-xy_c,rot) + xy_c + np.array([x,y])
        self.pose = self.pose + np.array([x,y,theta])
        
    def restore(self):
        self.pose = np.array([0,0,0])
        self.vertex_list = np.copy(self.vertex_list_original)
                                   
    def translate_guarded(self,dx,dy):
        #translate object in direction dx dy until collision with wall
        
        left_vertex = np.argmin(self.vertex_list[:,0])
        bottom_vertex = np.argmin(self.vertex_list[:,1])
        
        left_x = self.vertex_list[left_vertex,0]
        bot_y = self.vertex_list[bottom_vertex,1]
        
        # the point that will contact the wall moves like this as a function of time increasing.
        #wall_x = left_x + t*dx
        #wall_y = bot_y + t*dy
        
        if dx != 0:
            t1 = -(left_x - self.wall_x)/dx
        else:
            t1 = np.NaN
        
        if dy != 0:
            t2 = -(bot_y - self.wall_y)/dy
        else:
            t2 = np.NaN
        
        t = np.nanmin([t1,t2]) #earliest collision time
        
        #print t,t1,t2,t*dx,t*dy
        
        if t<0:
            raise ValueError('moving in direction with no collision')
        self.move(t*dx,t*dy,0)

    def r_bounding_circle(self):
        xy = self.vertex_list
        xy_c = self.get_pose()[0:2]
        dxy = (xy - xy_c)
        
        r = (dxy[:,0]**2 + dxy[:,1]**2)**.5
        return np.max(r) #max over the radius of all vertices
        
    def rotate_guarded(self,theta):
        if self.contact_torque() * theta < 0:
            return #trying to move into a contact.
        
        #xc,yc is center of object
        #dx_i,dy_i is the displacement from rotation center of vertex i
        
        xy = self.vertex_list
        xy_c = self.get_pose()[0:2]
        dxy = (xy - xy_c)
        
        r = self.r_bounding_circle()
        
        vertical_distance = xy_c[1]/r
        angle = np.arcsin(-vertical_distance)
        
        angles_intersection = np.array([angle, (np.pi/2-angle)+np.pi/2 ])
        
        angle_vertices = np.arctan2(dxy[:,1],dxy[:,0])
        
        def circle_dist_eq(frm,to,direction):
            d = circle_dist(frm,to,direction)
            if d < 1e-8:
                d = 2*np.pi
            return d
            
        if theta>0:
            dangles = [circle_dist_eq(frm,to,1) for frm in angle_vertices for to in angles_intersection]
            
        if theta<0:
            dangles = [-circle_dist_eq(frm,to,-1) for frm in angle_vertices for to in angles_intersection]
            
        #plot the path that the corners trace
        if False:
            plt.plot(xy_c[0],xy_c[1],'ko') #center
            
            for a in angles_intersection:
                plt.plot([xy_c[0], xy_c[0]+r*np.cos(a)],
                         [xy_c[1], xy_c[1]+r*np.sin(a)],lw=2)
            
            circ_patch =  matplotlib.patches.Circle(xy_c,r,fc='none')
            plt.gca().add_patch(circ_patch)
        
        t = np.array(dangles)/theta
        
        tp = t[t>=0]
        if len(tp) == 0:
            raise ValueError('no collision')
        t0 = np.nanmin(tp)
        
        self.move(0,0,theta*t0)
        
    def contact_vertices(self):
        contact_x = np.abs(self.vertex_list[:,0] - self.wall_x) < 1e-8
        contact_y = np.abs(self.vertex_list[:,1] - self.wall_y) < 1e-8
        
        #left-wall contacts, bottom-wall contacts
        return np.where(contact_x)[0], np.where(contact_y)[0]

    def contact_torque(self):
        leftcontact, bottomcontact = self.contact_vertices()
        xy_c = self.get_pose()[0:2]
        
        moment = np.float64(0.0)
        
        arms = self.vertex_list[bottomcontact,:] - xy_c
        for moments_due_to_bottom in arms[:,0]:
            moment += moments_due_to_bottom
        
        arms = self.vertex_list[leftcontact,:] - xy_c
        for moments_due_to_left  in arms[:,1]:
            moment += -moments_due_to_left
        
        return moment

    def intersects_bottom(self):
        left_vertex = np.argmin(self.vertex_list[:,0])
        bottom_vertex = np.argmin(self.vertex_list[:,1])
        
        left_x = self.vertex_list[left_vertex,0]
        bot_y = self.vertex_list[bottom_vertex,1]
        return bot_y < 0
        
def normang(a):
    while(a>np.pi):
        a = a - 2*np.pi
    while (a<-np.pi):
        a = a + 2*np.pi
    return a

def circle_dist(a1,a2,direction=1):
    a1 = normang(a1)
    a2 = normang(a2)
    
    while a1*direction > a2*direction:
        a2 += 2*np.pi*direction
    #make a2 come after a1 in the appropriate direction
    return (a2-a1)*direction
        

def plot_obj(obj,ax=None,kwline={},kwcontact={}):
    if ax is None:
        ax = plt.gca()
    
    i = range(obj.num_vertices) + [0]
    
    ax.plot(obj.vertex_list[i,0],obj.vertex_list[i,1],'b-',**kwline)
    
    ax.plot(obj.vertex_list[0,0],obj.vertex_list[0,1],'k*',**kwline) #to mark orientation
    
    leftcontact, bottomcontact = obj.contact_vertices()
    
    contacts = Set.union(Set(leftcontact),Set(bottomcontact))
    
    i = list(contacts)
    ax.plot(obj.vertex_list[i,0],obj.vertex_list[i,1],'r.',**kwcontact)
```

```python
def min_samples_from_state(T):
    return np.min( [ T[i].sum(axis=1).min() for i in range(len(T)) ] )
        
def min_samples_to_state(T):
    return np.min( [ T[i].sum(axis=0).min() for i in range(len(T)) ] )

def n_hits_state_in(T):
    c = np.zeros((len(states),))
    for i in range(len(actions)):
        c = c + T[i].sum(axis=1)
    return c
        
def n_hits_state_out(T):
    c = np.zeros((len(states),))
    for i in range(len(actions)):
        c = c + T[i].sum(axis=0)
    return c        
```

```python
def write_header(f):
    f.write("discount: 0.95\n")
    f.write("values: reward\n")
    f.write("states: %d\n"%(len(states)))
    f.write("actions: %d\n"%(len(actions)))
    f.write("observations: %d\n"%(len(observation_map)))
            

    
def write_start_state(f,distribution):
    f.write("start: ")
    for p in distribution:
        f.write("%f "%(p) )
    f.write("\n")
    
def write_start_state_uniform_subset(f,start_list):
    f.write("start include: ")
    for s in start_list:
        f.write("%d "%(s))
    f.write("\n")

def write_defaults(f):
    f.write("T:*:*:* 0\n") #sparse transitions
    f.write("R:*:*:*:* -1\n")
    f.write("O:*:*:0 1.0\n")
    
def write_transition(f, action,start,end,prob):
    f.write("T:%d:%d:%d %f\n"%(action,start,end,prob))
    
def write_observation(f,action,sink,observation,prob):
    s = "O:%d:%d:%d %f\n"%(action,sink,observation,prob)
    f.write(s)
    
def write_reward(f, action, start, end, obs, reward):
    s = "R:%d:%d:%d:%d %f\n"%(action,start,end,obs,reward)
    f.write(s)

def write_reward_action_endstate(f, action, end, reward):
    s = "R:%d:*:%d:* %f\n"%(action,end,reward)
    f.write(s)
    
def write_reward_action_fromstate(f, action, fromstate, reward):
    s = "R:%d:%d:*:* %f\n"%(action,fromstate,reward)
    f.write(s)    
    
def write_reward_endstate(f,end,reward):
    s = "R:*:*:%d:* %f\n"%(end,reward)
    f.write(s)  

def write_reward_fromstate(f,fromstate,reward):
    s = "R:*:%d:*:* %f\n"%(fromstate,reward)
    f.write(s)      
    
```

```python
obs = np.zeros(shape=(len(actions),len(states),len(observation_map)))
obs[:,:,observation_map['none']] = 1.0

#we have rewards depend only on (state,action), not next state and observation. But fake it so that the simulator is happy...

rws = np.zeros((len(actions),len(states),1,1))

from numpy.lib.stride_tricks import as_strided
stride = list(rws.strides)
stride[2] = stride[3] = 0
shape = list(rws.shape)
shape[2] = len(states)
shape[3] = len(observation_map)
rws = as_strided(rws, shape=shape, strides=stride)

rws[:,:,:,:] = -1

f = open('square.pomdp','w')

write_header(f)

start_dist = np.zeros((len(states)))
start_dist[start_states] = 1.0/np.sum(start_states)

#write_start_state(f,start_dist)

start_state_list = np.where(start_states)[0]
write_start_state_uniform_subset(f,start_state_list)

write_defaults(f)
for state_i,state in enumerate(states):
    po.set_pose(state)   #state we arrived in
    t = po.contact_torque()
    
    if t>1e-4:
        write_observation(f,actions_map['go_cw'],state_i,observation_map['ccw_torque'],.9)
        write_observation(f,actions_map['go_cw'],state_i,observation_map['none'],.1)
        
        obs[actions_map['go_cw'],state_i,observation_map['ccw_torque']] = .9
        obs[actions_map['go_cw'],state_i,observation_map['none']] = .1
        
        write_observation(f,actions_map['go_down'],state_i,observation_map['ccw_torque'],.8)
        write_observation(f,actions_map['go_down'],state_i,observation_map['none'],.2)

        obs[actions_map['go_down'],state_i,observation_map['ccw_torque']] = .8
        obs[actions_map['go_down'],state_i,observation_map['none']] = .2
    
    elif t<-1e-4:
        write_observation(f,actions_map['go_ccw'],state_i,observation_map['cw_torque'],.9)
        write_observation(f,actions_map['go_ccw'],state_i,observation_map['none'],.1)

        obs[actions_map['go_ccw'],state_i,observation_map['cw_torque']]=.9
        obs[actions_map['go_ccw'],state_i,observation_map['none']]=.1
        
        write_observation(f,actions_map['go_down'],state_i,observation_map['cw_torque'],.8)
        write_observation(f,actions_map['go_down'],state_i,observation_map['none'],.2)
        
        obs[actions_map['go_down'],state_i,observation_map['cw_torque']]=.8
        obs[actions_map['go_down'],state_i,observation_map['none']]=.2
        
    else: #basically no torque
        pass

#generate rewards        
for state_i,state in enumerate(states):
    po.set_pose(state)
    
    n_bottom_contacts = len(po.contact_vertices()[1])
    
    goal_n_contacts = 2 
    if n_bottom_contacts > 0 and n_bottom_contacts<goal_n_contacts: #basically if we have one contact, in this case.
        write_reward_fromstate(f,state_i,20*n_bottom_contacts)
        rws[:,state_i,:,:] = 20*n_bottom_contacts
        
    if n_bottom_contacts > 0:  #penalize going down if already in contact.
        write_reward_action_fromstate(f,actions_map['go_down'],state_i,-10)
        rws[actions_map['go_down'],state_i,:,:] = -10
        
    if n_bottom_contacts == goal_n_contacts:
        write_reward_action_fromstate(f,actions_map['go_none'],state_i,50)
        rws[actions_map['go_none'],state_i,:,:] = 50
        

#for (action_i,state_i,nstate_i) in zip(*np.where(Tprob>0)):
#    write_transition(f,action_i,state_i,nstate_i,Tprob[action_i,state_i,nstate_i]) 

for action_i in range(len(actions)):
    rows,cols = Tprob[action_i].nonzero()
    for j in range(Tprob[action_i].nnz):
        state_i,next_state_i = rows[j], cols[j]
        write_transition(f,action_i,state_i,next_state_i,Tprob[action_i][state_i,next_state_i])
        
f.close()
print 'done' 
```

```python
#this is the POMDP model that the solver is solving

Tprob_dense = [np.array(a.todense()) for a in Tprob]

model = pypomdp.model.POMDP(tps=Tprob_dense, 
                            ops=obs, 
                            rws=rws, 
                            #state_names=, 
                            action_names=actions_names, 
                            observ_names=observation_names )

model.discount = .95
model.values = 'reward' #reward or cost (sign of rewards)
```

```python
policy_filename = os.path.expanduser('~/out.policy')
policy_xml = ElementTree.parse(policy_filename)
alphavec_xml = policy_xml.getroot().find('AlphaVector') 

N_alpha_vecs = int(alphavec_xml.attrib['numVectors'])

assert int(alphavec_xml.attrib['vectorLength']) == len(states)
alphavecs = np.zeros((N_alpha_vecs,len(states)))

action_alphavec = np.zeros((N_alpha_vecs,),dtype=np.int)
for i,vec_xml in enumerate(alphavec_xml.getchildren()):
    action_alphavec[i] = vec_xml.attrib['action']
    alphavecs[i,:] = numpy.fromstring(vec_xml.text,sep=' ')

class AlphaVecPolicy(pypomdp.policy.Policy):
    def __init__(self,alphavecs,actions):
        self.alphavecs = alphavecs
        self.actions = actions
        assert len(self.actions) == len(self.alphavecs)

    def value(self,belief):
        return self.evaluate(belief)[1]
        
    def action(self,belief):
        return self.evaluate(belief)[0]
        
    def evaluate(self,belief):
        value = np.dot(self.alphavecs,belief)
        i = np.argmax(value)
        return self.actions[i], value[i]    
    
policy = AlphaVecPolicy(alphavecs,action_alphavec)
```

```python
class Simulator_Real():
    def __init__(self,start_state=None):
        self.po = po #PlanarPolygonObjectInCorner()
        if start_state is None:
            start_state = self.po.get_pose()
            while(self.po.intersects_bottom()):
                start_state = sample_continuous_state()[0]
         
        self.po.set_pose(start_state)
        self.current_state = self.po.get_pose()
        
    def simulate(self,action_i):
        actions[action_i]()
        self.current_state = self.po.get_pose()
        t = self.po.contact_torque() #contact as a result of the action
        
        reward = 0
        observ = observation_map['none']   
    
        if t>1e-4:
            if action_i == actions_map['go_cw']:
                #with .9
                observ = observation_map['ccw_torque']
            elif action_i == actions_map['go_down']:
                #with .8
                observ = observation_map['ccw_torque']
        
        elif t<-1e-4:
            if action_i == actions_map['go_ccw']:
                #with .9
                observ = observation_map['cw_torque']
            elif action_i == actions_map['go_down']:
                #with .8
                observ = observation_map['cw_torque']
                
        return reward,observ
```

```python
def belief_observation_update(last_action,this_observation,this_belief):
    state_likelihood = obs[last_action,:,this_observation].flatten()
    #state_likelihood[i]  is p(observed this_observation | world is in state i at this time)
    
    next_belief = state_likelihood * this_belief
    return next_belief/np.sum(next_belief)

def belief_transition_update(this_action,last_belief):
    #each row of Tprob[this_action] is a probability distribution
    return np.dot(last_belief,Tprob_dense[this_action])
    
```

```python
from collections import defaultdict

def marginalize_angle(belief):
    angle_belief = defaultdict(lambda: 0)
    
    for state_i in range(len(states)):
        _,_,theta = states[state_i]
        angle_belief[theta] += belief[state_i]
    
    v = np.sort(angle_belief.keys())
    return np.array([angle_belief[t] for t in v]),np.array(v)

max_height = np.max(states[:,1])
min_height = np.min(states[:,1])

bin_centers = np.linspace(min_height,max_height,8)
heights = states[:,1]

state_height_bins = [ [] for i in range(len(bin_centers))]

for state_i,state in enumerate(states):
    bin_i = np.argmin(np.abs(state[1]-bin_centers))
    state_height_bins[bin_i].append(state_i)
    
    
def marginalize_y(belief):
    return np.array([ np.sum ( belief[state_height_bins[i]] ) for i in range(len(bin_centers)) ]), bin_centers
    
        
```

```python
minangle = np.min(states[:,2])    
maxangle = np.max(states[:,2])


def plot_angle_belief(ax,belief):
    p,x = marginalize_angle(belief)
    w = .5/len(x) #center the bars
    ax.bar(x-w/2,p,0.5/len(x))
    ax.set_xlim(minangle,maxangle)    

def plot_height_belief(ax,belief):
    p,x = marginalize_y(belief)
    w = .5/len(x) #center the bars
    ax.bar(x-w/2,p,0.5/len(x))
    ax.set_xlim(min_height,max_height)
    
def make_uniform_height(belief):
    #takes a belief, returns another belief that has uniform height marginal.
    p,_ = marginalize_y(belief)
    
    rbelief = np.array(belief)
    for i in range(len(state_height_bins)):
        rbelief[state_height_bins[i]] /= p[i]*len(state_height_bins)
    
    return rbelief

def plot_belief_obj(ax,belief):
    s = np.random.choice(len(states),size=(50,),replace=True,p=belief)
    for state in s:
        po.set_pose(states[state])
        plot_obj(po,ax,{'alpha':.95,'color':'gray'})
        
```
